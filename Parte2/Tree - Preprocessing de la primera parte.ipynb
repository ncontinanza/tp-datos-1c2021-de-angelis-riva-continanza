```python
from sklearn.model_selection import train_test_split, GridSearchCV

from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import roc_curve, plot_roc_curve, auc,roc_auc_score, classification_report
from sklearn.metrics import confusion_matrix, plot_confusion_matrix, accuracy_score

import pandas as pd
import utils
import preprocessing

from matplotlib import pyplot as plt
import seaborn as sns

import graphviz
import dtreeviz.trees as dtreeviz
```

### Preprocesamiento del TP1

```python
df = preprocessing.preprocessing_base_parte_1(utils.feature_engineering(utils.get_data()))
```

Repartimos entre train y test

```python
X = df[df.columns.drop('tiene_alto_valor_adquisitivo')]
Y = df['tiene_alto_valor_adquisitivo']

X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=112)
```

Calcularemos, para cada altura entre 1 y 17, el mejor modelo obtenido para los parámetros:

* Criterio: Gini, Entropy
* Min muestras en hojas: 100, 200, 300, 400, 500

Para cada modelo entrenado se hace un kfold con k=20

```python
modelos = []

for altura in range(1, 18):
    print("Modelo con altura " + str(altura))
    model = DecisionTreeClassifier()
    params = {'criterion': ['gini','entropy'],
              'splitter': ['best'],
              'min_samples_leaf': [100, 200, 300, 400, 500],
              'max_depth': [altura]}
    gscv = GridSearchCV(model, params, scoring='accuracy', n_jobs=-1, cv=20, verbose=1)
    gscv.fit(X_train, y_train)
    print("El modelo obtuvo una clasificación media de: " + str(gscv.best_score_) + "\n")
    modelos.append(gscv.best_estimator_)
```

Graficaremos, para cada árbol obtenido en dicha búsqueda, el accuracy obtenido en train y en test. Para luego quedarnos con el modelo que tenga buen trade off complejidad-score

```python
accuracys = pd.DataFrame({"Altura": [], "Accuracy": [], "Tipo": []})
for modelo in modelos:
    modelo.fit(X_train, y_train)
    accuracys = accuracys.append({"Altura": modelo.get_depth(),
                      "Accuracy": accuracy_score(y_train, modelo.predict(X_train)),
                      "Tipo": "Train"}, ignore_index=True)
    accuracys = accuracys.append({"Altura": modelo.get_depth(),
                      "Accuracy": accuracy_score(y_test, modelo.predict(X_test)),
                      "Tipo": "Test"}, ignore_index=True)
```

```python
plt.figure(dpi=125)

sns.lineplot(data=accuracys,
             x = "Altura",
             y = "Accuracy",
             hue = "Tipo",
             palette="tab10", linewidth=2.5)

maxacc_idx = accuracys[accuracys["Tipo"] == "Test"]["Accuracy"].idxmax(axis=0)
maxacc_altura = accuracys.iloc[maxacc_idx]["Altura"]

plt.axvline(x = maxacc_altura, color="black", linestyle="--")

plt.title('Accuracy en train y test según la altura del árbol')
plt.xlabel("Altura del árbol", fontsize=14)
plt.ylabel("Accuracy del modelo", fontsize=14)
plt.xticks(range(1, 18))

plt.show()
```

Podemos ver en el gráfico que el mejor puntaje en test para el árbol fue cuando la altura estaba en 7, por lo que nos quedaremos con dicho árbol:

```python
arbol = modelos[6]
arbol.fit(X_train, y_train)
```

# El árbol

Graficaremos el árbol para verlo bien.

```python
viz = dtreeviz.dtreeviz(
    arbol,
    X_train,
    y_train,
    target_name='tiene_alto_valor_adquisitivo',
    feature_names=list(X.columns),
    class_names=list([0,1]),
    scale=1.5,
)

display(viz)
```

## Veamos ahora las demás métricas de nuestro modelo

### Roc Auc

```python
roc_auc_score(y_test, arbol.predict(X_test))
```

```python
fig, ax = plt.subplots(figsize=(5, 5), dpi=150)
plot_roc_curve(arbol, X_test, y_test, ax=ax)
plt.show()
```

El puntaje ROC dio bueno, aunque evidentemente podría ser mucho mejor. 

```python
print(classification_report(y_test, arbol.predict(X_test)))
```

```python
fig, ax = plt.subplots(figsize=(10,4), dpi=150)
plt.grid(False)
plot_confusion_matrix(arbol, X_test, y_test, cmap=plt.cm.Blues, ax=ax)
plt.show()
```

Podemos ver que tal cual nos pasó a nosotros en la primera entrega del tp, el recall para 1 es muy bajo, mientras que para 0 es muy alto. Si bien la diagonal sigue siendo dominante, la fila 1 está bastante mal, habría que ver cómo predecir mejor dichas instancias.
